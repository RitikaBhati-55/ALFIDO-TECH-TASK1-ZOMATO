```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv('zomato.csv')

# Print the first few rows to verify data
print(df.head())

# Data Cleaning
# Convert 'rate' to numeric (remove '/5' and handle non-numeric)
df['rate'] = df['rate'].astype(str).str.replace('/5', '').str.strip()
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Convert 'votes' to numeric
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')

# Convert 'approx_cost(for two people)' to numeric (remove commas)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str).str.replace(',', '').str.strip()
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

# Handle missing values: drop rows with NaN in key columns
df = df.dropna(subset=['rate', 'votes', 'location', 'cuisines', 'approx_cost(for two people)'])

# Visualization 1: Average Rating by Location (Top 10 Locations)
avg_rating_by_location = df.groupby('location')['rate'].mean().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='rate', y='location', data=avg_rating_by_location, palette='viridis')
plt.title('Top 10 Locations by Average Restaurant Rating')
plt.xlabel('Average Rating')
plt.ylabel('Location')
plt.tight_layout()
plt.savefig('avg_rating_by_location.png')  # Save the plot as an image
plt.show()

# Visualization 2: Distribution of Approximate Cost for Two People
plt.figure(figsize=(10, 6))
sns.histplot(df['approx_cost(for two people)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Approximate Cost for Two People')
plt.xlabel('Approx Cost (for two people)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('cost_distribution.png')  # Save the plot as an image
plt.show()

# Visualization 3: Top 10 Cuisines by Number of Restaurants
# Split cuisines if multiple and count occurrences
cuisines_series = df['cuisines'].str.split(', ').explode()
top_cuisines = cuisines_series.value_counts().nlargest(10).reset_index()
top_cuisines.columns = ['Cuisine', 'Count']

plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Cuisine', data=top_cuisines, palette='mako')
plt.title('Top 10 Cuisines by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.tight_layout()
plt.savefig('top_cuisines.png')  # Save the plot as an image
plt.show()

# Visualization 4: Pie Chart of Online Order Availability
online_order_counts = df['online_order'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(online_order_counts, labels=online_order_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Distribution of Restaurants with Online Order Option')
plt.tight_layout()
plt.savefig('online_order_distribution.png')  # Save the plot as an image
plt.show()

# Visualization 5: Top 10 Restaurants by Votes
top_restaurants_by_votes = df[['name', 'votes']].nlargest(10, 'votes')

plt.figure(figsize=(12, 8))
sns.barplot(x='votes', y='name', data=top_restaurants_by_votes, palette='rocket')
plt.title('Top 10 Restaurants by Number of Votes')
plt.xlabel('Votes')
plt.ylabel('Restaurant Name')
plt.tight_layout()
plt.savefig('top_restaurants_by_votes.png')  # Save the plot as an image
plt.show()

print("Visualizations saved as PNG files: avg_rating_by_location.png, cost_distribution.png, top_cuisines.png, online_order_distribution.png, top_restaurants_by_votes.png")
```
